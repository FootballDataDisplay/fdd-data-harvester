```python
# Import necessary libraries
import pandas as pd
import numpy as np  # Add this line
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load sample data (you'll replace this with your actual data later)
data = {
    'Team': ['Team A', 'Team B', 'Team C', 'Team D'],
    'Wins': [10, 8, 7, 11],
    'Losses': [6, 8, 9, 5],
    'Points_For': [350, 310, 280, 380],
    'Points_Against': [300, 320, 310, 270]
}
df = pd.DataFrame(data)
```

```python
# Display the first few rows of the data
print(df.head())
```

```python
# Basic statistical summary
print(df.describe())
```

```python
# Create a simple bar plot of wins by team
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Wins', data=df)
plt.title('Wins by Team')
plt.show()
```

```python
# Create a scatter plot of Points For vs Points Against
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Points_For', y='Points_Against', hue='Team', data=df)
plt.title('Points For vs Points Against by Team')
plt.show()
```

```python
# Correlation heatmap (only for numeric columns)
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numeric Features Only)')
plt.show()
```
